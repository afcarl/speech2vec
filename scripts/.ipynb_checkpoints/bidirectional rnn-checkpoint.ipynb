### Imports

```python
%matplotlib inline
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2

import config
from __future__ import division

import tensorflow as tf
from tensorflow.python.ops import rnn, rnn_cell
from tensorflow.python.ops import seq2seq
from tensorflow.examples.tutorials.mnist import input_data

import numpy as np

from tqdm import tqdm
```

### Load data

```python
from speech2vec.datasets import dsp_hw2

dataset = dsp_hw2()
X = dataset.fbank_delta
y = dataset.labels
```

### Learning Parameters

```python
# Learning parameters
learning_rate = 1e-3
nb_epochs = 1000
batch_size = 64

sample, timestep, feature = X.shape

hidden_dim = 256
keep = 0.8
```

### Clear tensorflow graph

```python
tf.reset_default_graph()
```

### Reuse variables

```python
tf.get_variable_scope().reuse_variables()
```

### Define cell

```python
# Define Cell
cell = tf.nn.rnn_cell.GRUCell(hidden_dim)
with tf.variable_scope('encoder'):
    en_cell = rnn_cell.MultiRNNCell([cell] * 2)
with tf.variable_scope('decoder'):
    de_cell = rnn_cell.MultiRNNCell([cell] * 2)
```

### Helper functions

```python
def xavier_init(fan_in, fan_out, constant=1): 
    """ Xavier initialization of network weights"""
    # https://stackoverflow.com/questions/33640581/how-to-do-xavier-initialization-on-tensorflow
    low = -constant*np.sqrt(6.0/(fan_in + fan_out)) 
    high = constant*np.sqrt(6.0/(fan_in + fan_out))
    return tf.random_uniform((fan_in, fan_out), 
                             minval=low, maxval=high, 
                             dtype=tf.float32)
```

### Encoder 

```python
# Input placeholder
x  = tf.placeholder(tf.float32, shape=[None, timestep, feature])

keep_prob = tf.placeholder(tf.float32)

# unpack sequence
encoder_inputs = tf.unpack(tf.transpose(x, perm=[1,0,2]))
encoder_outputs, encoder_state = tf.nn.rnn(en_cell, encoder_inputs, dtype=tf.float32)

code = tf.nn.relu(encoder_outputs[-1])
```

### Decoder

```python
code_drop = tf.nn.dropout( code, keep_prob )
decoder_inputs = [ code_drop ] + \
[ tf.placeholder(tf.float32, shape=[ None, hidden_dim ] ) for i in range( timestep -1 ) ]

def loop(prev, i):
    return prev # Output as input

decoder_outputs, decoder_state = seq2seq.rnn_decoder( decoder_inputs, tf.zeros_like(encoder_state), de_cell, loop_function = loop )

W_out = tf.Variable( xavier_init(hidden_dim, feature) )
b_out = tf.Variable( tf.zeros([ feature ] ) )

unpacked_reconstruction = [ tf.matmul( tf.nn.dropout( out, keep_prob ), W_out ) for out in decoder_outputs ]

x_ = tf.nn.relu( tf.transpose(tf.pack(unpacked_reconstruction), perm=[1, 0, 2]) )
```

### Cost & Optimizer

```python
cost = tf.reduce_mean(tf.reduce_sum(tf.square(x_-x),[1,2]))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
saver = tf.train.Saver()
```

### Training

```python
init = tf.initialize_all_variables()
with tf.Session() as sess:
    print "Initializing..."
    sess.run(init)

    for epoch in range(1, nb_epochs+1, 1):
        for batch_x, batch_x_hat in dataset.next_batch(batch_size=batch_size):
            sess.run( optimizer, feed_dict={x:batch_x, keep_prob: keep} )
        if epoch % 10 == 0:
            loss = sess.run(cost, feed_dict = { x: dataset.fbank_delta, keep_prob: 1. }) / timestep / feature
            print "Epoch",epoch,"Loss:", loss
    
        #save_path = saver.save(sess, "/tmp/dsp_hw2.ckpt")
        #print("Model saved in file: %s" % save_path)
```
